# Markov Inequality

Markov's Inequality is a fundamental theorem in probability theory that provides an upper bound on the probability that a non-negative random variable (X>=0) is greater than or equal to some positive constant value (a>0).



```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exponential distribution

```python
np.random.seed(42)
X = np.random.exponential(scale=1, size=1000)
a_values = np.linspace(0.1, 4, 50)

# Calculating P(X >= a) and E[X] / a
P_X_geq_a = [np.mean(X >= a) for a in a_values]
E_X_over_a = [np.mean(X) / a for a in a_values]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(a_values, P_X_geq_a, label='P(X >= a)', color='blue', linestyle='-', marker='o')
plt.plot(a_values, E_X_over_a, label='E[X] / a', color='red', linestyle='--')
plt.xlabel('a')
plt.ylabel('Probability')
plt.title("Markov's Inequality for X ~ Exponential(1)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)
X = np.random.exponential(scale=1, size=1000)
a_values = np.linspace(0.1, 10000, 5000)

# Calculating P(X >= a) and E[X] / a
P_X_geq_a = [np.mean(X >= a) for a in a_values]
E_X_over_a = [np.mean(X) / a for a in a_values]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(a_values, P_X_geq_a, label='P(X >= a)', color='blue', linestyle='-', marker='o')
plt.plot(a_values, E_X_over_a, label='E[X] / a', color='red', linestyle='--')
plt.xlabel('a')
plt.ylabel('Probability')
plt.title("Markov's Inequality for X ~ Exponential(1)")
plt.legend()
plt.grid(True)
plt.show()
```

There will never be a point at which the probability of a random variable being greater than some constant is greater than the expected value of the random variable divided by that constant.

## Uniform distribution

```python
np.random.seed(42)
X = np.random.uniform(low=0, high=1, size=1000)
a_values = np.linspace(0.1, 1, 50)

# Calculating P(X >= a) and E[X] / a
P_X_geq_a = [np.mean(X >= a) for a in a_values]
E_X_over_a = [np.mean(X) / a for a in a_values]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(a_values, P_X_geq_a, label='P(X >= a)', color='blue', linestyle='-', marker='o')
plt.plot(a_values, E_X_over_a, label='E[X] / a', color='red', linestyle='--')
plt.xlabel('a')
plt.ylabel('Probability')
plt.title("Markov's Inequality for X ~ Uniform(0, 1)")
plt.legend()
plt.grid(True)
plt.show()
```

## Poisson Distribution

```python
np.random.seed(42)
X = np.random.poisson(lam=1, size=1000)
a_values = np.linspace(0.1, 10, 50)

# Calculating P(X >= a) and E[X] / a
P_X_geq_a = [np.mean(X >= a) for a in a_values]
E_X_over_a = [np.mean(X) / a for a in a_values]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(a_values, P_X_geq_a, label='P(X >= a)', color='blue', linestyle='-', marker='o')
plt.plot(a_values, E_X_over_a, label='E[X] / a', color='red', linestyle='--')
plt.xlabel('a')
plt.ylabel('Probability')
plt.title("Markov's Inequality for X ~ Poisson(1)")
plt.legend()
plt.grid(True)
plt.show()
```
